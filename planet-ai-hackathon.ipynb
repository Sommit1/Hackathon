```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
```

```python
train_df = pd.read_csv("/kaggle/input/dataforhack-csv/Train_Data.csv")
test_df = pd.read_csv("/kaggle/input/dataforhack-csv/Test_Data.csv")
```

```python
train_df['age_group'] = train_df['age_group'].map({'Adult': 0, 'Senior': 1})
```

```python
train_df.dropna(subset=['age_group'], inplace=True)
```

```python
X = train_df.drop(columns=['SEQN', 'age_group'])
y = train_df['age_group']
X_test = test_df.drop(columns=['SEQN'])
```

```python
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)
X_test_imputed = imputer.transform(X_test)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
X_test_scaled = scaler.transform(X_test_imputed)
```

```python
model = RandomForestClassifier(
    n_estimators=500,
    max_depth=15,
    min_samples_split=3,
    min_samples_leaf=2,
    class_weight='balanced',
    random_state=42
)
model.fit(X_scaled, y)
```

```python
train_pred = model.predict(X_scaled)
train_f1 = f1_score(y, train_pred)
print("Training F1 Score:", round(train_f1 * 100, 2), "%")
```

```python
test_pred = model.predict(X_test_scaled)
```

```python
submission = pd.DataFrame({'age_group': test_pred})
submission.to_csv("submission.csv", index=False)
print("✅ submission.csv created for upload.")
```
